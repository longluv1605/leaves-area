```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
image_paths = []
data_path = 'datasets'
for root, dirs, files in os.walk(data_path):
    if 'spinach' in root:
        continue
    for file in files:
        if file.endswith('.jpg'):
            image_paths.append(os.path.join(root, file))

print(len(image_paths))
```

```python
image_shapes = []
for path in image_paths:
    image = cv2.imread(path)
    image_shapes.append(image.shape)
    
set(image_shapes)
```

```python
name = '/train/1_X2_1_1_22_11012025_jpg.rf.83b868d8080a44ceb2468b1d1ec80f6e'
image = cv2.imread('datasets/spinach/images' + name + '.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

mask = cv2.imread('datasets/spinach/masks' + name + '_mask.png', cv2.IMREAD_GRAYSCALE)

fig, axs = plt.subplots(1, 2, figsize=(8, 4))
axs[0].imshow(image)
axs[0].set_title("Image")
axs[1].imshow(mask, cmap='nipy_spectral')
axs[1].set_title("Mask")
for ax in axs:
    ax.axis('off')
plt.tight_layout()
plt.show()
```
